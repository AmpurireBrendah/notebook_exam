(b) Assumptions of linear regression 

1. Linearity: Relationship between target(dependent) and independent variables is linear for example y=b0+b1x where y is a target variabe and x is an independent variable.
2. Independence: The x(s) have to be identically independent thus use the same brand to test the model.
3. Homoscedasticity: There is a constant variance across all levels of independent variables in other words the values of x do not change since homo means one.
4. Normally distributed:The error follow a normal distribution (0,1).
5. No multicollinearity: Independent variables are not highly correlated.


(c)Train simple linear regression using scikit-learn

```python
# Define target variable (e.g., income-wages) and independent variable (e.g., Family_size)
x = census_data[['Family_size']]
y = census_data['Income-wages']
```

```python
# Splitting data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
print("Shape of the original dataset:",census_data.shape)
print("Shape of input - training set" ,x_train.shape)
print("Shape of output - training set" ,y_train.shape)
print("Shape of input - testing set", x_test.shape)
print("Shape of output - testing set",y_test.shape)
```

```python
# Create and train linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Get coefficients
coef = model.coef_[0] #B0
intercept = model.intercept_ #B1
# Generating the  equation
equation = f"Income-wages = {intercept:.2f} + {coef:.2f} * Family_size"
print(equation)
```

(d) Compute Mean Squared Error (MSE)

```python
# Make predictions
y_pred = model.predict(x_test)
```

```python
# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse:.2f}")
```

(e) Extract and interpret key outputs

From y=B0+B1x,
1.Coefficients(B1): Represent the change in target variable (y) for one unit change in independent variable(x).Focus on the slope and intercept.
2. Intercept(B0): Represents the expected value of target variable(y) when independent variable is zero.
3. R-squared: Measures goodness of fit (0-1).
4. MSE: Measures average squared difference between predictions and actual values.
5. Residual plots: Visualize differences between predictions and actual values.
